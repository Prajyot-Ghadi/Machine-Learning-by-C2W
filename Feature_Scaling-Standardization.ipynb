```python
import pandas as pd

from sklearn.preprocessing import StandardScaler

df = pd.read_csv('class_ipl_22-26_dataset.csv')
```

```python
df.head()
```

```python
df[['team1_runs']].head(20)
```

```python
scaler = StandardScaler()
df['team1_runs_scaled'] = scaler.fit_transform(df[['team1_runs']])

df[['team1_runs', 'team1_runs_scaled']].tail(74)
```

```python
scaler_wickets = StandardScaler()

df['team1_wickets_scaled'] = scaler.fit_transform(df[['team1_wickets']])

df[['team1_wickets', 'team1_wickets_scaled']].head()
```

```python
df[['team1_runs_scaled']].describe().round(2)
```

```python
df[['team1_wickets_scaled']].describe().round(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,5))

#plot the original team wickets
plt.subplot(1,2,1)
sns.histplot(df['team1_wickets'],kde=True )
plt.title("Distribution of Original Team1 Wickets")
plt.xlabel("Team1 Wickets")
plt.ylabel("Frerquency")


#plot for Scaled team1 wickets
plt.subplot(1,2,2)
sns.histplot(df['team1_wickets_scaled'], kde = True)
plt.title("Distribution of Scaled Team1 Wickets")
plt.xlabel("Scaled Team1 Wickets")
plt.ylabel("Frerquency")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

#plot the original team wickets
plt.subplot(1,2,1)
sns.histplot(df['team1_runs'],kde=True )
plt.title("Distribution of Original Team1 Runs")
plt.xlabel("Team1 Runs")
plt.ylabel("Frerquency")


#plot for Scaled team1 wickets
plt.subplot(1,2,2)
sns.histplot(df['team1_runs_scaled'], kde = True)
plt.title("Distribution of Scaled Team1 Runs")
plt.xlabel("Scaled Team1 Wickets")
plt.ylabel("Frerquency")

plt.tight_layout()
plt.show()
```
