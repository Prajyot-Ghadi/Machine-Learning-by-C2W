```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('class_ipl_22-26_dataset.csv')
df.head()
```

```python
X = df[['team1_runs']].tail(74)
X.head()
```

```python
X.index
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=['team1_runs'])

X_scaled.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.scatterplot(x=X.index, y=X['team1_runs'])
plt.title("Original team1_runs Distribution")
plt.xlabel("Index")
plt.ylabel("Team1 Runs")

plt.subplot(1,2,2)
sns.scatterplot(x=X_scaled.index, y=X_scaled['team1_runs'])
plt.title("Normalize team1_runs Distribution")
plt.xlabel("Index")
plt.ylabel("Normalized team1_runs")

plt.tight_layout()
plt.show()
```
