```python
print("Missing Values : ",df['winner'].isnull().sum())
print("Unique Values : ", df['winner'].nunique())
```

```python
df['winner'] = df['winner'].fillna('No Result')

df['winner'].isnull().sum()
```

```python
le = LabelEncoder()

df['winner_encoded'] = le.fit_transform(df['winner'])

df[['winner', 'winner_encoded']].tail(74)
```

```python
#Display Encoding Mapping

mapping = pd.DataFrame({
    'winner':le.classes_,
    'Encoded value' : range(len(le.classes_))
})
```
