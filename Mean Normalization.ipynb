```python
import pandas as pd

df = pd.read_csv('class_ipl_22-26_dataset.csv')
```

```python
df.head()
```

```python
X = df[['team1_runs']]
X.head()
```

```python
mean = X['team1_runs'].mean()
minimum = X['team1_runs'].min()
maximum = X['team1_runs'].max()

print("Mean : ", mean)
print("Minumum : ",minimum)
print("Maximum : ", maximum)
```

```python
X_mean = X.copy()

X_mean['team1_runs'] = (X_mean['team1_runs']-mean)/(maximum-minimum)

X_mean.head()
```

```python
comparision = pd.DataFrame({
    'Original' : X['team1_runs'],
    "Mean Normalized" : X_mean['team1_runs']
})

comparision.head()
```
