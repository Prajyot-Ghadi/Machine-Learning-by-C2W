```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('class_ipl_22-26_dataset.csv')
df.head()
```

```python
sns.relplot(
    data = df.tail(74),
    x = 'team1_runs',
    y = 'team2_runs',
    size = 'win_by_runs',
    kind = 'scatter'
    )
```

```python
sns.scatterplot(
    x='team2_runs',
    y='team1_runs',
    data = df.tail(74)
)
```

```python
sns.scatterplot(
    x = 'team2_runs',
    y='team2_wickets',
    data = df.tail(74)
)
```

```python
sns.countplot(
    x = 'winner',
    data = df.tail(74)

).tick_params(axis='x', rotation = 90)
```

```python
sns.barplot(
    x = 'season',
    y = 'team1_runs',
    data = df.tail(74)
).tick_params(axis='x', rotation = 90)
```

```python
sns.barplot(
    y = 'team1_runs',
    x = 'team1',
    data = df.tail(74),
    #errorbar = None
).tick_params(axis='x', rotation=90)
```

```python
sns.displot(df['team2_runs'].tail(74),
            kde = True)
```

```python
sns.displot(df['team2_wickets'].tail(74),
            kde=True)
```

```python
sns.histplot(df['winner'],
             kde = True).tick_params(axis ='x', rotation=90)
```

```python
sns.boxplot(
    x=df['team2_runs'].tail(74)
)
```

```python
df.tail(74).describe()
```

```python
sns.pairplot(df[['team1_runs', 'team2_runs']].tail(74))
```

```python
correlation = df.corr(numeric_only=True)
correlation
```

```python
sns.heatmap(
    correlation,
    annot=True,
    # cmap='Blues'
)
```

```python
sns.kdeplot(df['team1_runs'].tail(74),
            fill = True,
            bw_adjust = 0.5)
```

```python
import plotly.express as px

fig = px.scatter_3d(
    df.tail(74),
    x = 'team1_runs',
    y = 'team2_runs',
    z = 'win_by_runs',
    color = 'season'
)

fig.show()
```
