```python
import pandas as pd
```

```python
df = pd.read_csv('class_ipl_22-26_dataset_with_nulls.csv')
df.head()
```

```python
df['team1_wickets'].isnull().sum()
```

```python
df['team1_wickets'].head(20)
```

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=3, weights = 'distance')
df['team1_wickets'] = imputer.fit_transform(df[['team1_wickets']])
```

```python
df['team1_wickets'].isnull().sum()
```
