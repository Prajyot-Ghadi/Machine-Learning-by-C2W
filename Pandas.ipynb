```python
import pandas as pd
import matplotlib.pyplot as plt

ipl_df.std(numeric_only=True)

ipl_df = pd.read_csv('IPLPlayerStat.csv')

batting_avg_std = ipl_df['batting_avg'].std()

plt.figure(figsize=(12,6))
ipl_df['batting_avg'].plot(kind='hist', bins=15, edgecolor='black', alpha=0.7, density=True)
ipl_df['batting_avg'].plot(kind='kde', color='red',ax=plt.gca(), secondary_y = False)
plt.title('Distribution Of Batting Average (Histogram with KDE)')
plt.xlabel('Batting Average')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha= 'right')
plt.tight_layout()
plt.show()
```
