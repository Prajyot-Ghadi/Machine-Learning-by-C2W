```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

ipl = pd.read_csv('class_ipl_22-26_dataset.csv')
ipl.head()
```

```python
ipl.shape
ipl.columns
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="team1_wickets",
    y='team1_runs',
    data = ipl.tail(74)
)

plt.title('Team1 wickets vs Team1 Runs')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.scatterplot(
    x='team1',
    y='team2_runs',
    hue='winner',
    data = ipl.tail(74)
).tick_params(axis='x', rotation=90)

plt.title("Win by Wickets vs Win By Runs by Winner")
plt.show()
```

```python
multi_var_cols = ['team1_runs', 'team2_runs', 'win_by_runs', 'win_by_wickets', 'season', 'result_type']
multi_var_df = ipl[multi_var_cols].head(100)

sns.pairplot(multi_var_df, hue='result_type', diag_kind='kde')

plt.suptitle('Multivariate Analysis: Runs, Wickets, Seasons by Result types')
plt.show()
```

```python
numerical_cols = ['season', 'win_by_runs', 'win_by_wickets', 'team1_runs']

corr_df = ipl[numerical_cols]
corr_matrix = corr_df.corr()

plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
