## feature Selection

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("class_ipl_22-26_dataset.csv")
```

```python
df.head()
```

```python
data = df[['team1_runs','team2_runs', 'team1_wickets','team2_wickets', 'win_by_wickets']]
```

```python
data.head()
```

```python
correlation = data.corr()
correlation
```

```python
target_corr = correlation['team1_runs'].drop('team1_runs')
target_corr
```

```python
target_corr = target_corr.reindex(target_corr.sort_values(ascending=False).index)
target_corr
```

```python
def forword_selection(correlation_series):

    best_feature = correlation_series.idxmax()
    print("Selected Feature : ", best_feature)
    correlation_series = correlation_series.drop(best_feature)
    return correlation_series, best_feature
```

```python
selected = []
remaining = target_corr.copy()
remaining, feature = forword_selection(remaining)
selected.append(feature)
print("\n Selectec Feature : ", selected)
print("\n Remaining Features")
display(remaining)
```

```python
remaining, feature = forword_selection(remaining)
selected.append(feature)
print("\n Selectec Feature : ", selected)
print("\n Remaining Features")
display(remaining)
```

```python
print("Selected Feature Selections.ipynb")
selected_df = pd.DataFrame({
    "Selected Feature" : selected
})
display(selected_df)

print("\nRemaining Feature (Not Selected)")
display(remaining)
```

## PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("class_ipl_22-26_dataset.csv")
df.head()
```

```python
print("Shape : ", df.shape)
df.info()
```

```python
features = ['team1_runs', 'team1_wickets', 'team2_runs','team2_wickets', 'win_by_runs','win_by_wickets']

X = df[features]
X.head()
```

```python
import seaborn as sns 
corr = X.corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("correlation Matrix")
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[:5]
```

```python
X_scaled.mean().round(2)
```

```python
X_scaled.std()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(X_pca, columns=['PC','PC'])
pca_df.head()
```
