```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df = pd.read_csv('IPLPlayerStat.csv')
df.head()
```

```python
def player_form(avg):
    if avg <=10:
        return "Poor"
    elif avg <=15:
        return "Average"
    elif avg <=20:
        return "Good"
    elif avg <= 25:
        return "Excellent"
    else :
        return "Outstanding"

df["Player_Form"] = df["batting_avg"].apply(player_form)

df[["player", "batting_avg", "Player_Form"]].head()
```

```python
encoder = OrdinalEncoder(categories=[["Poor", "Average", "Excellent", "Good","Outstanding",]])

df["Player_Form_Encoded"] = encoder.fit_transform(df[["Player_Form"]]). astype(int)

df[["Player_Form", "Player_Form_Encoded"]].tail(30)
```
