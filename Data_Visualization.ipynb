```python
plt.plot(df['team1_runs'].tail(74))
plt.xlabel('Match Index')
plt.ylabel('Team 1 Runs')
plt.title('Team 1 Runs Trend (Last 74 Matches)')
plt.show()
```

```python
plt.plot(df['team2_runs'].tail(74), 'g--', label='Team 2 Runs')
plt.xlabel('Match Index')
plt.ylabel('Team 2 Runs')
plt.title('Team 2 Runs Trend (Last 74 Matches)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['team1_runs'].tail(74), label='Team runs', color='purple')
plt.plot(df['team1_wickets'].tail(74), label='Team Wickets', color='orange')
plt.xlabel('Runs')
plt.ylabel('wickets')
plt.title('Team run vs Team wickets')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['team1_runs'].tail(74), label='Team1 Runs', color='green')
plt.plot(df['team2_runs'].tail(74), label='Team2 Runs', color='maroon')
plt.xlabel('Match Index')
plt.ylabel('Runs')
plt.title('Team1 runs vs Team2 runs Trend (Last 74 Matches)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
match_indices = df.index[-74:]
plt.scatter(match_indices, df['team1_runs'].tail(74), c='blue', label='Team 1 Runs')
plt.scatter(match_indices, df['team2_runs'].tail(74), c='red', label='Team 2 Runs')
plt.xlabel('Match Index')
plt.ylabel('Runs')
plt.title('Team 1 Runs and Team2 Runs per Match (last 74 Matches)')
plt.legend()
plt.show()
```

```python
plt.hist(df['team2_wickets'].tail(74),bins=10)
plt.xlabel('wickets')
plt.ylabel('Number of Matches')
plt.title('Distribution of Team 2 wickets')
plt.show
```

```python
toss_decisions = df['toss_decision'].value_counts()
plt.figure(figsize=(8,6))
plt.bar(toss_decisions.index, toss_decisions.values, color=['blue', 'red'])
plt.xlabel('Toss Decision')
plt.ylabel('Number of Matches')
plt.title('Distribution of Toss Decisions')
plt.show()
```

```python
team_wins = df['winner'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(team_wins.index, team_wins.values, color=['red', 'blue'])
plt.xlabel('Winninf team')
plt.ylabel('Number of Matches')
plt.title('Distribution of Winning Teams')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.boxplot(df['team2_runs'])
plt.ylabel("Team 2 Runs")
plt.title('Distributuion of Team 2 Runs')
plt.show()
```

```python
result_type_counts = df['winner'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(result_type_counts, labels=result_type_counts.index, pctdistance=1, radius=2, autopct='%1.1f%%', rotatelabels=True, startangle=90)
plt.title("Distribution of Match Result Types")
plt.xticks(rotation=90)
plt.show()
```
