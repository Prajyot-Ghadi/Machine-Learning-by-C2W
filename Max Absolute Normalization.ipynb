```python
import pandas as pd
from sklearn.preprocessing import MaxAbsScaler
```

```python
df = pd.read_csv('class_ipl_22-26_dataset.csv')
df.head()
```

```python
X = df[['team1_runs', 'team2_runs']]
X.head()
```

```python
scaler = MaxAbsScaler()

X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=['team1_runs','team2_runs'])
X_scaled.head()
```

```python
comparision = pd.DataFrame({
    'Original' : X['team1_runs'],
    'Max Absolute Scaled' : X_scaled['team1_runs']
})

comparision.tail(74)
```

```python
print("Original Data")
print(X.describe())

print("\nScaled Data")
print(X_scaled.describe())
```
